```python
def transform_to_opposite(transformed,presite):
    trans = np.copy(transformed)
    num = presite.shape[0]
    for i in tqdm(range(num)):
        #print (np.sum(trans==presite[i,1]))
        trans[trans==presite[i,1]] = - presite[i,0]
    return trans
```
